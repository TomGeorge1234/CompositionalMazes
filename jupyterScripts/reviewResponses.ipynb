# Repeat originals with longer time

## Open loop 

```python
N=50
roomsize = 5
xcen = np.linspace(0,roomsize,N+1)[:-1]
xcen += (xcen[1] - xcen[0]) / 2
ycen = np.array([0.1]*N)
centres = np.vstack((xcen,ycen)).T

params = { 
          'mazeType'            : 'loop',  
          'stateType'           : 'gaussianThreshold', 
          'movementPolicy'      : 'windowsScreensaver',  
          'doorsClosed':False,
          'speedScale':0.16, 
          'roomSize'            : roomsize,   
          'dt'                  : 0.003,     
          'centres'             : centres, 
          }
```

```python
agent =  MazeAgent(params)
```

```python
agent = MazeAgent(loadFromFileCalled="openLoop_60mins")
```

```python
agent.runRat(trainTime=60,saveEvery=0.5)
```

```python
agent.saveToFile(name="openLoop_60mins")
```

```python
plotter = Visualiser(agent)
fig, ax = plotter.plotFeatureCells(centresOnly=True, textlabel=False,onepink=True)
saveFigure(fig,'openloop',specialLocation="../figures/paperFigures/reviewerresponses/long_open_maze.svg")

fig, ax = plotter.plotM(whichM='W',colorbar=False,time=30)
ax.axis('on')
for spine in ax.spines.values():
    spine.set_color('C1')
    spine.set_linewidth(2)
ax.set_xticks([])
ax.set_yticks([])
saveFigure(fig,'W',specialLocation='../figures/paperFigures/reviewerresponses/long_open_STDP.svg')

fig, ax, cb, cax = plotter.plotM(whichM='M')
ax.axis('on')
for spine in ax.spines.values():
    spine.set_color('C0')
    spine.set_linewidth(2)
cb.set_ticks([0])
cb.set_ticklabels([""])
# cax.tick_params(color='darkgrey',width=2)
ax.set_xticks([])
ax.set_yticks([])
saveFigure(fig,'M',specialLocation='../figures/paperFigures/reviewerresponses/long_open_SR.svg')

fig, ax = plotter.plotMAveraged(time=30)
saveFigure(fig,'Mav',specialLocation='../figures/paperFigures/reviewerresponses/longer_open_Av.svg')

fig, ax = plotter.plotMetrics(x_ticks=[0,30,60])
saveFigure(fig,'Mvarerr',specialLocation='../figures/paperFigures/reviewerresponses/long_open_metrics.svg')
```
